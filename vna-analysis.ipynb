```python
import csv
import numpy as np
import itertools
import matplotlib.pyplot as plt
import colormap as cmaps
import pandas as pd
```

```python
colnames = ['freq[Hz]', 're:Trc1_S21', 'im:Trc1_S21','re:Trc2_S11', 'im:Trc2_S11']

data = pd.read_csv('data/test1.csv',delimiter=';',skiprows=2) #names=colnames, delimiter=';')
#data = pd.read_csv('data/insulated/emptycavitys11.csv',delimiter=';',skiprows=2) #names=colnames, delimiter=';')
data = data.iloc[: , :-1]
data.head()

#data.shape
```

```python
plt.rcParams["figure.figsize"] = (9,9)
plt.rcParams["font.size"] = 20
plt.rcParams["axes.titlesize"] = 20
#plt.rcParams["axes.labelsize"] = 16
plt.rcParams["legend.fontsize"] = 20

data["freq[Hz]"].plot()
plt.show()
data["re:Trc1_S21"].plot(label='S21_RE')
plt.legend()
plt.show()
data["im:Trc1_S21"].plot(label='S21_IM')
plt.legend()
plt.show()

#data["re:Trc1_S11"].plot()
#plt.show()
#data["im:Trc1_S11"].plot()
#plt.show()

data['S21_db'] = np.log(abs(data['re:Trc1_S21']))
data.plot(x='freq[Hz]', y='S21_db')
plt.ylabel('dB')
plt.xlabel('Frequency [GHz]')
plt.show()
```
